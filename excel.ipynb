2) Describe the functions and usefulness of 'Freeze Panes' and 'Split Panes'
in Excel.
->
### 1. Freeze Panes

**Function:**
Freeze Panes keeps specific rows or columns visible while you scroll through the rest of the worksheet.

**Usefulness:**
It helps you view headings or labels (like column titles or row names) at all times, making large datasets easier to read.

**Example:**

* Freezing the **top row** keeps column headers visible when scrolling down.
* Freezing the **first column** keeps row labels visible when scrolling right.

---

### 2. Split Panes

**Function:**
Split Panes divides the worksheet window into separate, scrollable sections.

**Usefulness:**
It allows you to view and compare different parts of the same worksheet at the same time.

**Example:**

* Splitting the sheet lets you scroll through data at the bottom while still viewing data at the top.
* Useful when comparing values from two distant areas in a large worksheet.



3) Explain the difference between inserting a new row and inserting a new
column in Excel. Can you insert multiple rows or columns at once?
->
### Difference between inserting a new row and a new column in Excel

#### Inserting a new row

* **What happens:** A new horizontal row is added.
* **Where it appears:** Above the selected row.
* **Used for:** Adding new records or entries.
* **Example:**
  If you select **row 5** and insert a row, the new blank row appears above row 5.

#### Inserting a new column

* **What happens:** A new vertical column is added.
* **Where it appears:** To the **left** of the selected column.
* **Used for:** Adding new fields or categories.
* **Example:**
  If you select **column C** and insert a column, the new column appears before column C.

---

### Inserting multiple rows or columns

**Yes, you can insert multiple rows or columns at once.**

**How:**

* Select multiple rows or columns first.
* Right-click and choose **Insert**, or use **Home â†’ Insert â†’ Insert Sheet Rows/Columns**.

**Example:**
Selecting **3 rows** and inserting will add **3 new rows** above them.
Selecting **2 columns** and inserting will add **2 new columns** to the left.




4)What are logical functions in Excel? Provide examples of at least two
logical functions and their applications.
->
### Logical functions in Excel

**Logical functions** in Excel are used to make decisions by testing conditions. They return results based on whether a condition is **TRUE or FALSE** and are commonly used for analysis, grading, validation, and automation.

---

### Examples of logical functions and their applications

#### 1. IF function

**Purpose:**
Tests a condition and returns one value if the condition is TRUE and another if it is FALSE.

**Syntax:**
`=IF(logical_test, value_if_true, value_if_false)`

**Example & application:**
`=IF(A1>=50, "Pass", "Fail")`

* Used in **exam results** to determine pass or fail based on marks.

---

#### 2. AND function

**Purpose:**
Checks multiple conditions and returns TRUE only if **all** conditions are TRUE.

**Syntax:**
`=AND(condition1, condition2, ...)`

**Example & application:**
`=AND(A1>=50, B1>=75)`

* Used for **eligibility checks**, such as passing both theory and practical exams.




5)Discuss the purpose of 'XLOOKUP' and how it differs from the traditional
'VLOOKUP' function.
->
### Purpose of **XLOOKUP**

**XLOOKUP** is a modern lookup function in Excel used to **search for a value in one range and return a corresponding value from another range**. It was introduced to overcome many limitations of older lookup functions.

**Basic syntax:**
`=XLOOKUP(lookup_value, lookup_array, return_array, [if_not_found], [match_mode], [search_mode])`

**Example:**
`=XLOOKUP(A2, A:A, C:C)`
Looks for the value in **A2** within column A and returns the matching value from column C.

---

### How XLOOKUP differs from VLOOKUP

| Feature          | XLOOKUP                         | VLOOKUP                             |
| ---------------- | ------------------------------- | ----------------------------------- |
| Lookup direction | Can look **left or right**      | Can look **only to the right**      |
| Column insertion | Not affected                    | Breaks if columns are added/removed |
| Exact match      | Default                         | Requires FALSE argument             |
| Error handling   | Built-in `if_not_found`         | Needs IFERROR                       |
| Match types      | Exact, approximate, wildcard    | Limited                             |
| Range selection  | Separate lookup & return ranges | Uses column index number            |

---

### Example comparison

* **VLOOKUP:**
  `=VLOOKUP(A2, A:C, 3, FALSE)`
  Returns value from the 3rd column to the right.

* **XLOOKUP:**
  `=XLOOKUP(A2, A:A, C:C, "Not Found")`
  More flexible, clearer, and safer.
